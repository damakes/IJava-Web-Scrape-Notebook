```java
import org.jsoup.Jsoup;
import org.jsoup.nodes.Document;
import org.jsoup.nodes.Element;
import org.jsoup.select.Elements;

import java.io.FileWriter;
import java.io.BufferedWriter;
import java.io.IOException;
```

```java
String url = "https://fi.wikipedia.org/wiki/Luettelo_Suomen_kuntien_koordinaateista";
String filePath = "csv/Suomen_kuntien_koordinaatit.csv";
```

```java
Document doc = Jsoup.connect(url).get(); 
doc
```

```java
Elements rows = doc.select("table.wikitable tbody tr");
rows
```

```java
rows.select("td");
```

```java
try (BufferedWriter writer = new BufferedWriter(new FileWriter(filePath))) {

    writer.write("Location, Latitude, Longitude");
    writer.newLine();


    for (int i = 0; i < rows.size(); i++) {
        Element row = rows.get(i);
        Elements cols = row.select("td, th");

        if (i > 0 && cols.size() >= 2) {
            String location = cols.get(0).text();
            String coordinates = cols.get(1).text();

            writer.write(location + ", " + coordinates);
            writer.newLine();
            
            System.out.println(location + ", " + coordinates);
        }
    }
} catch (IOException e) {
    System.err.println("Error writing to file: " + e.getMessage());
}
```
